Para poder obter resultados máis lexibles, podemos facelo cun bucle que vaia percorrendo as frases (`sentences`) e, dentro destas, as palabras(`words`), e dándonos información máis entendible.

Ademáis, imos xerar o dicionario neste paso, ademáis

```python
#comezamos cun dicionario baleiro
dic_polar={}
#variable f de número de frase
f = 0
#lista p para número de palabras por frase
p = []
#lista negativo para número de palabras negativas por frase
negativo=[]
#lista positivo para número de palabras positivas por frase
positivo=[]
#lista neutro para número de palabras neutras por frase
neutro=[]
#lista punt_frase para las puntuaciones por frase
punt_frase=[]
#variable total para la puntuación total
total=0

```

```python
#por cada frase
for frase in texto_procesado_gl.sentences:
  #donde iba el f+=1 (fallo)
  #f += 1
  #print(f)
  #por cada palabra
  p.insert(f,0)
  negativo.insert(f,0)
  positivo.insert(f,0)
  neutro.insert(f,0)
  punt_frase.insert(f,0)

  for palabra in frase.words:

    if (palabra.upos != 'PUNCT'):
      p[f] += 1
    #visualiza por pantalla a información:
    print("Posición: ", palabra.id," \t Palabra: ", palabra.text, "\t Lema: ", palabra.lemma, "\t Etiqueta: ", palabra.upos,  "\t Head: ", palabra.head,"\t Feat: ", palabra.feats, "\t Dep: ", palabra.deprel,sep="")

    #imos dar valores positivos negativos e neutros a cada palabra e xerar así o dicionario

#imos dar valores positivos negativos e neutros a cada palabra e xerar así o dicionario
    if palabra.text == 'morriñoso':
      dic_polar['morriñoso'] = -1
      negativo[f] += 1
    elif palabra.text ==  'vedreiro':
      dic_polar['vedreiro'] = -1
      negativo[f] += 1
    elif palabra.text == 'noxo':
      dic_polar['noxo'] = -1
      negativo[f] += 1
    elif palabra.text == 'famento':
      dic_polar['famento'] = -1
      negativo[f] += 1
    elif palabra.text == 'manso':
      dic_polar['manso'] = 1
      positivo[f] += 1
  f += 1




for i in range(f):
  neutro[i] = p[i] - ((negativo[i]+positivo[i]))


for r in range(f):
  punt_frase[r] = positivo[r] - negativo[r]
  total += punt_frase[r]

print(positivo)
print(negativo)
print(neutro)
print(p)
print(punt_frase)
print(total)
```
